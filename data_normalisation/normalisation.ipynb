## Normalización de Datos

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
clase1 = pd.read_csv('./data/clase_1.csv', delimiter= ';')
clase2 = pd.read_csv('./data/clase_2.csv', delimiter = ';')
clase3 = pd.read_csv('./data/clase_3.csv', delimiter= ';')
clase4 = pd.read_csv('./data/clase_4.csv', delimiter = ';')
claustro = pd.read_csv('./data/claustro.csv', delimiter = ';')
```

### Alumnos completos

```python
#Juntando
alumnos_ds = pd.merge(clase1,clase2, how= 'outer')
alumnos_ds['Vertical'] = 'Data Science'
alumnos_fs = pd.merge(clase3, clase4, how= 'outer')
alumnos_fs['Vertical'] = 'Full Stack'
alumnos = pd.merge(alumnos_ds, alumnos_fs, how='outer')

#Limpiando
alumnos.reset_index(inplace= True)
alumnos.rename(columns= {'Proyecto_FullSatck': 'ProyectoFullStack'}, inplace = True)


alumnos.drop(columns='index', inplace= True)
```

```python
alumnos.head()
```

```python
alumnos.to_csv('alumnos_completo.csv', index = False)
```
